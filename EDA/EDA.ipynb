```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
book_rating = pd.read_csv("Books_rating.csv")
books_data = pd.read_csv("books_data.csv")
```

```python
# Merge two datasets
books = pd.merge(book_rating, books_data, on='Title', how='inner')
books.dropna(subset=['Title'], inplace=True)
```

```python
books = books.drop(['image', 'previewLink', 'infoLink'], axis=1)

# Data Overview
books.head()
```

```python
# Check missing values
count_stats = books.isnull().sum().to_frame('count_na')
count_stats['count_na_pct'] = count_stats['count_na'] / len(books)
count_stats
```

```python
# Fill missing value for price and ratingsCount with median
books['ratingsCount'].fillna(value=books['ratingsCount'].median(), inplace=True)
books['Price'].fillna(value=books['Price'].median(), inplace=True)

# Drop missing value for other columns.
books.dropna(inplace=True)
```

```python
# Check missing values
books.isnull().sum()
```

```python
books['categories'] = books['categories'].str.extract(r'\'(.*)\'')
books['authors'] = books['authors'].str.extract(r'\'(.*)\'')
```

```python
# Number of books written by Authors
top_authors = books.groupby('authors')['Id'].nunique().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_authors.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Number of Books Written by Authors (Top 10)')
plt.ylabel('Author')
plt.xlabel('Number of Books')
plt.grid(axis='x', linewidth=0.5)
plt.gca().invert_yaxis()
plt.show()
```

```python
# Top 10 genre
top_categories = books.groupby('categories')['Id'].nunique().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_categories.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.title('Top 10 Genres')
plt.ylabel('')
plt.axis('equal')
plt.legend(top_categories.index, title="Genres", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

```python
# Top 10 Books with highest number of ratings
top_rating = books.groupby('Id')['ratingsCount'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_rating.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Top 10 Books with highest number of ratings')
plt.ylabel('Book Id')
plt.xlabel('Number of Ratings')
plt.grid(axis='x', linewidth=0.5)
plt.gca().invert_yaxis()
plt.show()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vader = SentimentIntensityAnalyzer()
data = books.sample(20000)
def get_sentiment_category(compound_score):
    if compound_score >= 0.05:
        return 'positive'
    elif compound_score < -0.05:
        return 'negative'
    else:
        return 'neutral'

data['compound'] = data['review/text'].str.lower().apply(lambda review: vader.polarity_scores(review)['compound'])
data['Sentiment'] = data['compound'].apply(get_sentiment_category)
```

```python
# Distribution of Sentiments
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
sentiment_counts = data['Sentiment'].value_counts()
ax[0].bar(sentiment_counts.index, sentiment_counts.values, color='skyblue', edgecolor='black')
ax[0].set_title('Distribution of Sentiments')
ax[0].set_xlabel('Sentiment')
ax[0].set_ylabel('Count')
ax[0].grid(axis='y', linewidth=0.5)

ax[1].pie(sentiment_counts.values, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
ax[1].set_title('Distribution of Sentiments')
ax[1].legend(sentiment_counts.index, title="Sentiments", loc="best")
ax[1].axis('equal')

plt.tight_layout()
plt.show()
```

```python
# Distribution of Sentiments (Histogram)
negative = data[data['compound'] < 0]['compound']
neutral = data[data['compound'] == 0]['compound']
positive = data[data['compound'] > 0]['compound']

# Plot each category with a different color and label
plt.figure(figsize=(10, 6))
plt.hist(negative, bins=20, color='wheat', alpha=0.7, label='Negative')
plt.hist(neutral, bins=20, color='salmon', alpha=0.7, label='Neutral')
plt.hist(positive, bins=20, color='skyblue', alpha=0.7, label='Positive')

plt.title('Distribution of Compound Sentiment Scores')
plt.xlabel('Compound Score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linewidth=0.5)
plt.show()
```

```python
# Number of Positive/Negative/Neutral Review on books(Top30)
top_positive_reviews = data[data['Sentiment'] == 'positive']['Id'].value_counts().head(30)
top_negative_reviews = data[data['Sentiment'] == 'negative']['Id'].value_counts().head(30)
top_neutral_reviews = data[data['Sentiment'] == 'neutral']['Id'].value_counts().head(30)

fig, ax = plt.subplots(2, 2, figsize=(12, 10))

# Positive Reviews
top_positive_reviews.plot(kind='bar', ax=ax[0, 0], color='lightblue', edgecolor='black')
ax[0, 0].set_title('Top 30 Books by Number of Positive Reviews')
ax[0, 0].set_ylabel('Number of Positive Reviews')
ax[0, 0].set_xticklabels(ax[0, 0].get_xticklabels(), rotation=90)
ax[0, 0].grid(axis='y', linewidth=0.5)

# Negative Reviews
top_negative_reviews.plot(kind='bar', ax=ax[0, 1], color='lightblue', edgecolor='black')
ax[0, 1].set_title('Top 30 Books by Number of Negative Reviews')
ax[0, 1].set_ylabel('Number of Negative Reviews')
ax[0, 1].set_xticklabels(ax[0, 1].get_xticklabels(), rotation=90)
ax[0, 1].grid(axis='y', linewidth=0.5)

# Neutral Reviews
top_neutral_reviews.plot(kind='bar', ax=ax[1, 0], color='lightblue', edgecolor='black')
ax[1, 0].set_title('Top 30 Books by Number of Neutral Reviews')
ax[1, 0].set_ylabel('Number of Neutral Reviews')
ax[1, 0].set_xticklabels(ax[1, 0].get_xticklabels(), rotation=90)
ax[1, 0].grid(axis='y', linewidth=0.5)

ax[1, 1].axis('off')

plt.tight_layout()
plt.show()
```
